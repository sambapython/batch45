```python
import random
names=["anil","ashok","ramesh","madhan"]
ages=[29,31,28,27,26]
sals=[12000,23000,45000,60000]
rows=[]
for i in xrange(1000):
    row="%s,%s,%s,%s\n"%(i,random.choice(names),random.choice(ages),random.choice(sals))
    rows.append(row)
f=open("data.csv","w")
f.writelines(rows)
f.close()
    
```
